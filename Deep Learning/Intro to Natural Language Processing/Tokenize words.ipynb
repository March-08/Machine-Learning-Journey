It is helpful to tokenize words in senteces, like 
- "I love you" : 1,2,3
- "I love football" : 1,2,4

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
sentences = [
  "I love my Dog",
  "I love my Cat",
  "My name is Marcello and I am a Computer Scientists"
]
```

In Tokenizer "num_word" to encode only n words, the most commons.
"oov_token" is a token to raplace words not known by the tokenizer.
Notice the tokinezire cleans punctuations, upper cases etc...

```python
tokenizer = Tokenizer(num_words=100, oov_token="<OOV>")
tokenizer.fit_on_texts(sentences)
tokenizer.index_word
```

Now let's convert these sentences to sequences using these tokens

```python
sequences = tokenizer.texts_to_sequences(sentences)
```

If we want all the sequences with same lenght in order to feed the Neural Network we should add Padding

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

Lets ad the zeros at the end (post). We can choose a maxlen, and the sentences longer will be trunctated at the beginning (pre)

```python
padded = pad_sequences(sequences,padding ="post",maxlen = 6 , truncating="pre")
padded
```

```python
test_sentences = [
  "I am a pro",
  "I love videogames"
]
```

```python
test_sequences = tokenizer.texts_to_sequences(test_sentences) #remember 1 is <OOV>
test_sequences
```

```python
padded_test_sequences = pad_sequences(test_sequences,padding ="post",maxlen = 6 , truncating="pre")
padded_test_sequences
```
